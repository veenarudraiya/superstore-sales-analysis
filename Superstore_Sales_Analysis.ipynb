```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_quantity = df['Quantity'].sum()

print("üí∞ Total Sales: ‚Çπ", round(total_sales, 2))
print("üèÜ Total Profit: ‚Çπ", round(total_profit, 2))
print("üì¶ Total Quantity Sold:", total_quantity)
```

```python
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values('Sales', ascending=False)
print(category_summary)
```

```python
region_summary = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values('Sales', ascending=False)
print(region_summary)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a modern visual style
sns.set(style="whitegrid", palette="pastel")

print("‚úÖ Visualization libraries imported successfully!")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum, ci=None)
plt.title("Total Sales by Category", fontsize=14)
plt.ylabel("Sales (in ‚Çπ)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum, ci=None)
plt.title("Sales by Region", fontsize=14)
plt.ylabel("Sales (in ‚Çπ)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Sub-Category', y='Profit', estimator=sum, ci=None)
plt.title("Total Profit by Sub-Category", fontsize=14)
plt.xticks(rotation=45)
plt.ylabel("Profit (in ‚Çπ)")
plt.show()
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by month
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend", fontsize=14)
plt.ylabel("Sales (in ‚Çπ)")
plt.xlabel("Month")
plt.show()
```
